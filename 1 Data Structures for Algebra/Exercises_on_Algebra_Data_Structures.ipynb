```python
import torch
import tensorflow as tf
```

```python
#PYTORCH
v_pt = torch.tensor([[25],[2],[-3],[-23]])
v_pt
```

```python
v_pt.shape
```

```python
v_pt_t = v_pt.T
v_pt_t
```

```python
#TENSORFLOW
v_tf = tf.Variable([[25],[2],[-3],[-23]])
v_tf
```

```python
v_tf.shape
```

```python
v_tf_t = tf.transpose(v_tf)
v_tf_t
```

```python
import numpy as np
```

```python
A = np.array([[3, 4, 5], [6, 7, 8]])
A
```

```python
A.shape
```

```python
b = np.array([[1],[2],[3]])
b
```

```python
b.shape
```

```python
#buat matrix A_pt dengan ukuran 2x3 dan vector b_pt berukuran 3x1 menggunakan pytorch
import torch
```

```python
A_pt = torch.tensor([[9, 10, 11], [12, 13, 14]])
A_pt
```

```python
b_pt = torch.tensor([[11], [12], [13]])
b_pt
```

```python
#buat matrix A_tf dengan ukuran 2x3 dan vector b_tf berukuran 3x1 menggunakan tensorflow
import tensorflow as tf
```

```python
A_tf = tf.Variable([[15, 16, 17], [18, 19, 20]])
A_tf
```

```python
b_tf = tf.Variable([[14], [15], [16]])
b_tf
```

```python
#tampilkan ukuran matrix A dan vector b
print(A.shape)
print(b.shape)
print(A_pt.shape)
print(b_pt.shape)
print(A_tf.shape)
print(b_tf.shape)
```

```python
#kalikan matrix A dengan vector b
np.dot(A, b)
```

```python
#kalikan matrix A_pt dengan vector b_pt
torch.matmul(A_pt, b_pt)
```

```python
#kalikan matrix A_tf dengan vector b_tf
tf.matmul(A_tf, b_tf)
```

```python
#buat matrix B dengan ukuran 3x3 menggunakan numpy
B = np.array([[3, 6, 5], [9, 1, 2], [8, 7, 2]])
B
```

```python
#buat matrix B_pt dengan ukuran 3x3 menggunakan pytorch
B_pt = torch.tensor([[9, 10, 11], [12, 13, 14], [3, 5, 8]])
B_pt
```

```python
#buat matrix B_tf dengan ukuran 3x3 menggunakan tensorflow
B_tf = tf.Variable([[2, 5, 0], [6, 1, 0], [8, 8, 4]])
B_tf
```

```python
#kalikan matrix A dengan matrix B
np.dot(A, B)
```

```python
#kalikan matrix A_pt dengan matrix B_pt
torch.matmul(A_pt, b_pt)
```

```python
#kalikan matrix A_tf dengan matrix B_tf
tf.matmul(A_tf, B_tf)
```

```python
#buktikan apakah perkalian matrix bersifat komutatif? (AB=BA)?
np.dot(A, B)
```

```python
np.dot(B, A)
```

```python
#Jika di lihat A(2x3)*B(3x3) berhasil dan B(3x3)*A(2x3)tidak bisa karena kolom B dan baris A tidak sama sehingga muncul eror
#kesimpulan : Tidak Komutatif!
```

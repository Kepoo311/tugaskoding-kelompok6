```python
#buat visualisasi dari persamaan berikut
#y = 2x+1 dan 3x-y = 5
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,25,50)
```

```python
y1 = 2*x + 1
y2 = 3*x - 5
```

```python
fig, ax = plt.subplots()
plt.title('Visualisasi Persamaan Garis')
plt.xlabel('x')
plt.ylabel('y')
ax.set_xlim([-10, 25])
ax.set_ylim([-10, 50])
ax.plot(x, y1, c='green', label='y = 2x + 1')
ax.plot(x, y2, c='brown', label='y = 3x - 5')
ax.scatter(6, 13, color='red')
plt.axvline(x=6, color='black', linestyle='--')
plt.axhline(y=13, color='black', linestyle='--')

```

SOAL SUB BAB 1

```python
import numpy as np
import torch
import tensorflow as tf
```

```python
# Buat matrix X (2x3) dengan NumPy
X = np.array([[1, 2, 3],
              [4, 5, 6]])
X
```

```python
# Buat matrix X_pt (2x3) dengan PyTorch
X_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]], dtype=torch.float32)
X_pt
```

```python
# Buat matrix X_tf (2x3) dengan TensorFlow
X_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]], dtype=tf.float32)
X_tf
```

```python
scalar = 2

# Perkalian
mul_np = X * scalar
mul_pt = X_pt * scalar
mul_tf = X_tf * scalar
print (mul_np)
print (mul_pt)
print (mul_tf)
```

```python
# Penjumlahan
add_np = X + scalar
add_pt = X_pt + scalar
add_tf = X_tf + scalar
print (add_np)
print (add_pt)
print (add_tf)
```

```python
# Pengurangan
sub_np = X - scalar
sub_pt = X_pt - scalar
sub_tf = X_tf - scalar
print (sub_np)
print (sub_pt)
print (sub_tf)
```

SOAL SUB BAB 2

```python
# Buat matrix Y (2x3) dengan NumPy
Y = np.array([[1, 2, 3],
              [4, 5, 6]])
Y
```

```python
# Buat matrix Y_pt (2x3) dengan PyTorch
Y_pt = torch.tensor([[1, 2, 3],
                     [4, 5, 6]], dtype=torch.float32)
Y_pt
```

```python
# Buat matrix Y_tf (2x3) dengan TensorFlow
Y_tf = tf.constant([[1, 2, 3],
                    [4, 5, 6]], dtype=tf.float32)
Y_tf
```

```python
# NumPy
res_np = X * Y

# PyTorch
res_pt = X_pt * Y_pt

# TensorFlow
res_tf = X_tf * Y_tf

print (res_np)
print (res_pt)
print (res_tf)
```
